### Finding the best model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from keras_tuner import RandomSearch
from keras_tuner import Objective
```

```python
def build_model(hp):
    model = Sequential()
    for i in range (hp.Int('num_layers_large',1,5)):
        model.add(Dense(units=hp.Int('units_lar'+str(i),
                                     min_value=500,
                                     max_value=750,
                                     step=40),
                       kernel_initializer=(hp.Choice('weight_lar'+str(i),['glorot_uniform','glorot_normal','he_uniform','he_normal','normal','uniform'])),
                       activation='relu'))
        model.add(Dropout(hp.Choice('dropout_l'+str(i),[0.01,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.450,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.850,0.9,0.95])))
    for j in range (hp.Int('num_layers_mid',1,2)):
        model.add(Dense(units=hp.Int('units_mid'+str(j),
                                     min_value=100,
                                     max_value=500,
                                     step=22),
                       kernel_initializer=(hp.Choice('weight_mid'+str(j),['glorot_uniform','glorot_normal','he_uniform','he_normal','normal','uniform'])),
                       activation='relu'))
        model.add(Dropout(hp.Choice('dropout_m'+str(j),[0.01,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.450,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.850,0.9,0.95])))
    model.add(Flatten())
    model.add(Dense(units=hp.Int('units_small',
                                 min_value=25,
                                 max_value=100,
                                 step=18),
                    kernel_initializer=(hp.Choice('weight_small',['glorot_uniform','glorot_normal','he_uniform','he_normal','normal','uniform'])),
                    activation='relu'))
    model.add(Dropout(hp.Choice('dropout_s',[0.01,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.450,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.850,0.9,0.95])))
    model.add(Dense(1,kernel_initializer=(hp.Choice('weight_final',['glorot_uniform','glorot_normal','he_uniform','he_normal','normal','uniform'])),activation='linear'))
    model.compile(optimizer=Adam(learning_rate=(hp.Choice('learning_rate',[1e-2,75e-4,5e-3,25e-4,1e-3,75e-5,5e-4,25e-5,1e-4,75e-6,5e-5,25e-6,1e-5,75e-7,5e-6,25e-7,1e-6,75e-8,5e-7]))),
                  loss='mean_squared_error',
                  metrics=['mean_absolute_percentage_error'])
    return model
```

```python
tuner=RandomSearch(build_model,
                  objective=Objective('val_mean_absolute_percentage_error',direction='min'),
                  max_trials=20,
                  executions_per_trial=1,
                  directory='project',
                  project_name='energy_production2')
```

```python
stoping=EarlyStopping(patience=5,verbose=1,min_delta=0.01)
```

```python
tuner.search(X_train,Y_train,validation_data=(X_valid,Y_valid),epochs=6,batch_size=40)
```

```python
tuner.results_summary(1)
```

```python
best_hp=tuner.get_best_hyperparameters()[0]
model=tuner.hypermodel.build(best_hp)
model.build(input_shape=[0, 5, 2, 2, 2])
```

```python
model.summary()
```

```python
stoping=EarlyStopping(monitor='val_mean_absolute_percentage_error',patience=6, verbose=1, min_delta=0.0001,restore_best_weights=True)
```

```python
rec=model.fit(X_train,Y_train,validation_data=(X_valid,Y_valid),epochs=100,batch_size=40,callbacks=[stoping])
```

```python
model.save('model.h5')
```
